# Import libraries 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder

import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("C:/Users/HP/Desktop/ExcelR/Data Science/Python/Assignments/Association Rules/my_movies.csv")
data
```

```python
data = data.iloc[:,5:]
data
```

# Apriori Algorithm 

```python
frequent_movies = apriori(data, min_support=0.1, use_colnames=True)
frequent_movies
```

```python
rules = association_rules(frequent_movies, metric="lift", min_threshold=0.7)
rules
```

```python
rules.sort_values('lift',ascending = False)
```

#####  An leverage value of 0 indicates independence. Range will be [-1 1]
A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```

```python
rules[['support','confidence']].hist()
```

```python
rules[['support','confidence','lift']].hist()
```

```python
plt.scatter(rules['support'], rules['confidence'])
plt.show()
```

```python
sns.scatterplot('support', 'confidence', data=rules, hue='antecedents')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```
